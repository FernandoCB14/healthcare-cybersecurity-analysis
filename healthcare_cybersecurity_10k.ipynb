```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("healthcare_cybersecurity_10k.csv")
df.info()
df.head()
df["Severity"].value_counts()
df["CVSS_Score"].describe()
sns.countplot(data=df, x="Severity")
plt.show()

df['Attack_Vector'].value_counts()
df["Weakness"].value_counts().head(10)

df["Attack_Vector"].value_counts()
sns.boxplot(data=df, x="Severity", y="CVSS_Score")
plt.show()

top_weaknesses = df["Weakness"].value_counts().head(10)

sns.barplot(
    x=top_weaknesses.values,
    y=top_weaknesses.index
)

plt.show()
```
